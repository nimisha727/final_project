```python
# Dependencies:
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Locate the file
file = 'Resources/CAERS_ASCII_2004_2017Q2.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Preview of the Dataframe
df.head(2)
```

```python
# Drop all rows with missing information:
df = df.dropna(how='any')
```

```python
# Verify dropped rows:
#df.count()
```

```python
# Renaming the Columns:
df.columns = ['Report Num','Report Date','Event Start Date',
              'Product role','Producy Name','Industry','Industry Name',
              'Age','Age Unit','Gender','Outcomes','Side Effects']
df.head()                  
```

```python
df.describe()
```

```python
from pymongo import MongoClient

client = MongoClient()
col = client['test']['test']

data = df.to_dict(orient='records')
col.insert_many(data)
```

```python
#Create connection to SQL Database
from sqlalchemy import create_engine
engine = create_engine('sqlite:///Resources/adv_effects.sqlite', echo=False)
```

```python
#Create new table in SQL DB
df.to_sql('adv_effects', con=engine)
```

```python
engine.execute('SELECT * FROM adv_effects').fetchall()
```
